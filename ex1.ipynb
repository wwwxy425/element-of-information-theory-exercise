1.随机生成1000个均值为0，方差为1的高斯分布的样本，可以使用np.random.normal。
2.画出这1000个高斯分布样本的直方图，注意：纵轴要显示样本数量，可以使用matplotlib.pyplot.hist。
3.在样本直方图上，叠加画出标准高斯分布的概率密度函数，通过改变概率密度的幅度，使其与样本直方图大致重合。

```python
import numpy as np
import matplotlib.pyplot as plt


xx=np.random.normal(0.0,1.0,1000)

def gauss(x):
    return(1/(np.sqrt(2*np.pi)))*np.exp(-(1/2)*(x**2))

XX=np.linspace(-3,3,100)
yy=gauss(XX)


plt.hist(x=xx,bins=20,density=True)
plt.plot(XX,yy,color="r")
plt.xlabel("random number")
plt.ylabel("probabity density")
plt.show()
```

熵、相对熵、联合熵、互信息的Python实现:

```python
import numpy as np
import math



def log(x):
     return np.where(x>0,np.log2(x),0)

#熵
def entropy(px):
    # return the entropy of the random variable 'x' with probability distribution 'px'
    # e.g. px = [1/2, 1/2,  0]
    return -np.sum(px*log(px))

#相对熵
def relative_entropy(px, qx):
    # return the relative entropy of the random variable 'x' with probability distribution 'px' and 'qx'
    # e.g. qx = [5/36, 1/3, 19/36], px = [1/2,1/2,0]
    return np.sum(px * log(px)-px*log(qx))

#联合熵
def joint_entropy(px,py):
    # return the joint entropy of the random variable 'x' and 'y' with joint p.d. 'pxy'
    # e.g. qxy = [[1/18, 1/12],
    # [1/18, 5/18], [5/18, 1/4]]
    return -np.sum(np.outer(px,py)*log(np.outer(px,py)))

#互信息
def mutual_infomation(px, py, pxy):
    # return the mutual information of I(x;y)
    return np.sum(pxy*(log(pxy)-log(np.outer(px,py))))



px = np.array([1/2, 1/2, 0])  
qx = np.array([5/36, 1/3, 19/36])  
qy = np.array([7/18, 11/18])  
qxy = np.array([[1/18, 1/12], [1/18, 5/18], [5/18, 1/4]])
print("熵: ", entropy(px), "相对熵: ", relative_entropy(px, qx), "联合熵: ", joint_entropy(qx, qy), "互信息: ", mutual_infomation(qx, qy, qxy))


```
